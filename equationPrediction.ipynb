```python
import torch
```

```python
#@title Fonksiyon tanımı y = f(x)
def f(in_tensor):
    katsayilar = torch.tensor([0.8, -0.5, 0.7], dtype=torch.float).unsqueeze(1)
    return in_tensor.mm(katsayilar) -0.2
```

```python
# Veri seti oluşturalım

x = torch.rand((100,3),dtype=torch.float)
y = f(x)

x_train = x[0:80, :]
x_test = x[80:100, :]
y_train = y[0:80, :]
y_test = y[80:100, :]
```

```python
#Sistemi oluşturalım

class Fonksiyon(torch.nn.Module):
  def __init__(self):
    super(Fonksiyon, self).__init__()
    self.carpan_katsayilar = torch.nn.Parameter(torch.rand((3, 1), dtype=torch.float))
    self.sabit_katsayi = torch.nn.Parameter(torch.rand((1, 1), dtype=torch.float))

  def forward(self, x):
    return x.mm(self.carpan_katsayilar) + self.sabit_katsayi

model = Fonksiyon()
```

```python
loss_fonksiyonu = torch.nn.MSELoss(reduction='sum')

optimizer = torch.optim.Adam(model.parameters(), lr=0.1)
```

```python
for i in range(200):
  y_tahmin = model(x_train)

  loss = loss_fonksiyonu(y_tahmin, y_train)
  
  optimizer.zero_grad()

  loss.backward()

  optimizer.step()
```
